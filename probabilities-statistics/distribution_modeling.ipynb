```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)

# normal data
n_normal = 1000
X_normal = np.random.normal(loc=0, scale=1, size=(n_normal, 2))

# anomalies (far from normal distribution)
n_anomaly = 50
X_anomaly = np.random.uniform(low=6, high=8, size=(n_anomaly, 2))

# combine
X = np.vstack([X_normal, X_anomaly])

# labels: 0 = normal, 1 = anomaly
y = np.hstack([np.zeros(n_normal), np.ones(n_anomaly)])

plt.plot(X[y == 0, 0], X[y == 0, 1], 'o', alpha=0.5)
plt.plot(X[y == 1, 0], X[y == 1, 1], 'o', alpha=0.5)
plt.show()
```

```python
# plot distribution of X
plt.hist(X[:, 0], bins=30, alpha=0.5)
plt.hist(X[:, 1], bins=30, alpha=0.5)
plt.show()

plt.show()
```

```python
# z-score
z_scores = np.abs((X - X.mean(axis=0)) / X.std(axis=0))
outliers = (z_scores > 3).any(axis=1)

plt.plot(X[outliers, 0], X[outliers, 1], 'o', color='red', label='outliers', alpha=0.5)
plt.plot(X[~outliers, 0], X[~outliers, 1], 'o', color='blue', label='normal', alpha=0.5)
plt.legend()
plt.show()
```
