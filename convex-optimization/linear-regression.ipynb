```python
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

def generate_dataset(n: int, noise_std: float = 100.0) -> tuple[np.ndarray, np.ndarray]:
    a = np.random.uniform(low=-100, high=100)
    x = np.linspace(start=0, stop=100, num=n)
    noise = np.random.randn(n) * noise_std
    y = a * x + noise
    return x, y

x, y = generate_dataset(n=100)
px.scatter(x=x, y=y)
```

```python
from typing import Literal, Optional
import numpy as np

def mse(y_hat: np.ndarray, y: np.ndarray) -> None:
    return np.mean((y_hat - y) ** 2)

class LinearRegression:
    def __init__(self, input_features: int, output_features: int) -> None:
        self.input_features = input_features
        self.output_features = output_features

        self.theta = np.random.randn(input_features, output_features)

    def _reset_theta(self) -> None:
        self.theta = np.random.randn(self.input_features, self.output_features)

    def _get_close_form_solution(self, x: np.ndarray, y: np.ndarray) -> np.ndarray:
        x = np.asarray(x, dtype=np.float64).reshape(-1, self.input_features)
        y = np.asarray(y, dtype=np.float64).reshape(-1, self.output_features)
        theta = np.linalg.solve(x.T @ x, x.T @ y)
        return theta

    def forward(self, x: np.ndarray) -> np.ndarray:
        x = np.asarray(x, dtype=np.float64).reshape(-1, self.input_features)
        return x @ self.theta

    def fit(self, x: np.ndarray, y: np.ndarray, method=Literal["close_form", "iterative"], epochs: Optional[int] = 10, learning_rate: Optional[float] = 1e-5) -> None:
        if method == "close_form":
            self._reset_theta()
            self.theta = self._get_close_form_solution(x=x, y=y)
        else:
            raise NotImplementedError("chill bro")

linear_regression = LinearRegression(input_features=1, output_features=1)
linear_regression.fit(x=x, y=y, method="close_form")
```

```python
y_hat = linear_regression.forward(x=x)

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=x,
    y=y,
    mode="markers",
))

fig.add_trace(go.Scatter(
    x=x,
    y=y_hat.ravel(),
    mode="lines",
))

fig.show()
```
